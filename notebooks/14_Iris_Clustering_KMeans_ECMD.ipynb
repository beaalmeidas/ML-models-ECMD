```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Loading the dataset
iris = load_iris()

# Taking only the data, no labels
iris_data = iris.data
```

```python
# In the context of the Iris dataset, the number of clusters is often chosen
# as 3 because the dataset contains three flower species: Setosa, Versicolor and Virginica
model = KMeans(n_clusters=3, random_state=42)

# Treinar o modelo
model.fit(iris_data)
```

```python
# Getting the labels of the clusters
cluster_labels = model.labels_

# Getting the coordinates of the centroids
centroid_coordinates = model.cluster_centers_
```

```python
# Inertia in the context of the K-Means algorithm is a metric used to evaluate
# the quality of the clusters formed. It measures the degree of cohesion of
# the points within each cluster, meaning how close the data points are to their
# respective centroids (the mean point of the cluster).
inertia = model.inertia_

print("CLUSTER INERTIA:")
print(inertia)
```

```python
# MEASURING THE SILHOUETTE
silhouette_avg = silhouette_score(iris_data, cluster_labels)

# The Silhouette coefficient evaluates how well the points are grouped
# into clusters and varies from -1 to 1. Values ​​closer to 1 indicate better clustering.

print("SILHOUETTE COEFICIENT:")
print(silhouette_avg)
```

```python
# Visualizar clusters em 2D (usando as duas primeiras features)
plt.scatter(iris_data[:, 0], iris_data[:, 1], c=cluster_labels, cmap='viridis')
plt.scatter(centroid_coordinates[:, 0], centroid_coordinates[:, 1], s=300, c='red', label='Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-MEANS CLUSTERS')
plt.legend()
plt.show()
```
