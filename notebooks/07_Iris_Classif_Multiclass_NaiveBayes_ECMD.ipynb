```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, confusion_matrix, accuracy_score
```

```python
# Loading the dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names) # This only gets the study features

# Loading the classification column
iris_df['target'] = iris.target

# Printing info about the dataset
print(iris_df)
print("\n")
print(iris_df.info())
```

```python
# Defining the columns that are studied to make the classification
x = np.array(iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])

# Defining the classification column
y = np.array(iris_df[['target']])

# Splitting the data into the training(70%) and validation(30%) sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
print("{0:0.2f}% in the training set".format((len(x_train)/len(iris_df.index)) * 100))
print("{0:0.2f}% in the testing set".format((len(x_test)/len(iris_df.index)) * 100))
```

```python
model = GaussianNB()
model.fit(x_train, y_train.ravel())
```

```python
# Predicting
y_pred = model.predict(x_test)
```

```python
# MEASURING MODEL WEIGHTED PRECISION
print("MODEL WEIGHTED PRECISION SCORE:")
precision_score(y_test, y_pred, average='weighted')
```

```python
model_conf_matrix = confusion_matrix(y_test, y_pred)
print("MODEL CONFUSION MATRIX:")
print(model_conf_matrix)
```

```python
model_accuracy_score = accuracy_score(y_test, y_pred)
print(f"MODEL ACCURACY SCORE: {model_accuracy_score:.2f}")
```
