```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Generating fake data

np.random.seed(42)

# Number of samples
n_samples = 100

# Columns
temperature = np.random.uniform(15, 35, n_samples)
humidity = np.random.uniform(30, 90, n_samples)
n_people = np.random.randint(1, 10, n_samples)

# Using a linear function to create energy consumption data
error = np.random.normal(0, 5, n_samples)
energy_consumption = 10 + 1.5 * temperature + 0.5 * humidity + 2 * n_people + error

energy_data = pd.DataFrame({
    'Temperature': temperature,
    'Humidity': humidity,
    'Number_Of_People': n_people,
    'Consumption': energy_consumption
})
```

```python
# Splitting
x = energy_data[['Temperature', 'Humidity', 'Number_Of_People']]
y = energy_data['Consumption']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initializing model and predicting
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

```python
# EVALUATING THE MODEL
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MODEL MSE:")
print(mse)
print("\n")
print("MODEL R² SCORE:")
print(r2)
```

```python
# PLOTTING A GRAPH
# This graph compares the real consumption values with the predicted consumption values
plt.scatter(y_test, y_pred)
plt.xlabel('REAL CONSUMPTION')
plt.ylabel('PREDICTED CONSUMPTION')
plt.title('REAL VS PREDICTED CONSUMPTION')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2)
plt.show()
```
