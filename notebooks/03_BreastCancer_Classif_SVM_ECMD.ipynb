```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
breast_cancer_data = pd.read_csv("breast-cancer.csv", encoding="utf-8")

print(breast_cancer_data.info())
print("\n")
breast_cancer_data.head(5)
```

```python
breast_cancer_data.tail(5)
```

```python
# Converting 'nucleo_descoberto' e 'classe' to numeric
label_encoder = LabelEncoder()
breast_cancer_data['nucleo_descoberto'] = label_encoder.fit_transform(breast_cancer_data['nucleo_descoberto'])
breast_cancer_data['classe'] = label_encoder.fit_transform(breast_cancer_data['classe'])
```

```python
print(breast_cancer_data.info())
```

```python
# Defining which column the model is supposed to study for the predictions
x = breast_cancer_data.drop(columns=['classe'])

# Defining which column the model is trying to predict
y = breast_cancer_data['classe']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Escalation is necessary because if the features are in different scales, the
# SVM algorithm might struggle with weighing their importance among the dataset
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Training
model = SVC()
model.fit(x_train_scaled, y_train)

# Predicting
y_pred = model.predict(x_test_scaled)
```

```python
# PRINTING THE PRECISION AND RECALL SCORES
print(classification_report(y_test, y_pred))
```
